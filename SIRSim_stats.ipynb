```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#compute mean
import statistics
```

```python
# read data
data = pd.read_csv("result1")
```

```python
data
```

```python
df_end_of_runs = data[data["Infected"] == 0]
df_end_of_runs
```

```python
print(statistics.mean(df_end_of_runs["Susceptible"]), statistics.stdev(df_end_of_runs["Susceptible"]))
print(statistics.mean(df_end_of_runs["Removed"]), statistics.stdev(df_end_of_runs["Removed"]))
print(statistics.mean(df_end_of_runs["Cycle"]),statistics.stdev(df_end_of_runs["Cycle"]))

```

```python
df_end_of_runs_divide = df_end_of_runs[["Susceptible","Infected","Removed"]]/10010
df_end_of_runs_divide
```

```python
df_index = df_end_of_runs["Cycle"]
df_index_list = list(df_index.index)
df_index_list = [0] + df_index_list
print(df_index_list)
```

```python
list_data_tables = []
def split_table():
    for i in range(len(df_index_list)-1):
        #print(i)
        #print(i, df_index_list[i])
        if i == 0:
            list_data_tables.append(data[df_index_list[i]:df_index_list[i+1]+1])
        else:
            list_data_tables.append(data[df_index_list[i]+1:df_index_list[i+1]+1])
            
```

```python
split_table()
```

```python
for i in range(len(list_data_tables)-1):
    print(list_data_tables[i])
```

```python
len(list_data_tables)
```

```python
def plot(list_data_tables):
    for i in range(len(list_data_tables)):
        figure = plt.figure()
        figure.canvas.set_window_title('SIR model')

        inf_line, =plt.plot(list_data_tables[i]["Infected"], label='I(t)')

        sus_line, = plt.plot(list_data_tables[i]["Susceptible"], label='S(t)')

        rec_line, = plt.plot(list_data_tables[i]["Removed"], label='R(t)')
        plt.legend(handles=[inf_line, sus_line, rec_line])
        
        
```

```python
plot(list_data_tables)
```
